# SIR Model Simulation of Disease Spread

This project models how infectious diseases spread using the SIR (Susceptible-Infected-Recovered) model.

The population is divided into:
- Susceptible (S)
- Infected (I)
- Recovered (R)

The model is based on differential equations and solved using Python.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Total population
N = 1000

# Initial conditions
I0 = 1      # initial infected
R0 = 0      # initial recovered
S0 = N - I0 - R0  # initial susceptible

# Parameters
beta = 0.3   # infection rate
gamma = 0.1  # recovery rate

# Time (days)
t = np.linspace(0, 100, 100)

# SIR differential equations
def sir_model(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt

# Initial vector
y0 = S0, I0, R0

# Solve ODE
solution = odeint(sir_model, y0, t, args=(N, beta, gamma))
S, I, R = solution.T

# Plot results
plt.figure(figsize=(10,6))
plt.plot(t, S, 'b', label='Susceptible')
plt.plot(t, I, 'r', label='Infected')
plt.plot(t, R, 'g', label='Recovered')

plt.xlabel('Time (days)')
plt.ylabel('Population')
plt.title('SIR Model Simulation')
plt.legend()
plt.grid()

plt.show()
```

## Observations

- The susceptible population decreases over time
- The infected population rises to a peak and then falls
- The recovered population steadily increases

## Conclusion

The infection rate (beta) controls how fast the disease spreads, while the recovery rate (gamma) determines how quickly people recover.
